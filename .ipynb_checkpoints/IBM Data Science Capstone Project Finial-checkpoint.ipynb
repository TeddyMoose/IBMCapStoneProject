##### Below we define a function which will output the venue type from json data for each venue in the reutrned json data from FourSquare



```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

#### Below is a defined function which will run through all the post codes storing the output to a single dataframe
It will allso simutainsly combine the resutls with there appropiate postcode

```python
def getNearbyVenues(names, latitudes, longitudes, radius=2000):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print('Processing...' +name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

### One Hot encoding
One hot encoding takes the unique values for venue catagory and transpose them in to columns. Once the columns are applied it then add a 1 to the corosopding column that the venue is in.


```python
# one hot encoding
sheffield_onehot = pd.get_dummies(sheffield_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
sheffield_onehot['Neighborhood'] = sheffield_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [sheffield_onehot.columns[-1]] + list(sheffield_onehot.columns[:-1])
sheffield_onehot = sheffield_onehot[fixed_columns]

sheffield_onehot.head(10)
```

**With the Sheffield_onehot datafarme we create a new dataframe sheffield_grouped which groups all the venues into there postcodes, then a mean is applied to the fill in the values for each venue catagory. This creates a % frenquency for each venue catagory in each area**

```python
sheffield_grouped = sheffield_onehot.groupby('Neighborhood').mean().reset_index()
sheffield_grouped
```

**In order to make this easier to view we run through each area and print a table containing the top 5 venues, there fequnecy % and catagory name**
This can be seen in the output cell below

```python
num_top_venues = 5

for hood in sheffield_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = sheffield_grouped[sheffield_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

### Create a function to return Nth number of most common venues

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

### The above function is the applied to our dataset, to create a new dataframe **neighborhoods_venues_sorted** which shows the top N venues by post code

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = sheffield_grouped['Neighborhood']

for ind in np.arange(sheffield_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(sheffield_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## K Means Clustering
In the cell below build and run the kmeans object, then out put the array containg the predict k means labels.

```python
kclusters = 10

sheffield_grouped_clustering = sheffield_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(sheffield_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

#### Now the kmeans labels are joined to the dataframe so we align out cluster labels with there corosponding post codes.
This means we can see which postcode belong to which clusters. This tabel is croweded and hard to read so in the following steps we will filter them and map them to make them easier to view.

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

sheffield_merged = sheffPC

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
sheffield_merged = sheffield_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Postcode')

sheffield_merged.head() # check the last columns!
```

## Mapping Results
The results from the kmeans clustering is plotted on the map of sheffield. With colors corosponding to the cluster number

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.jet(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(sheffield_merged['Latitude'], sheffield_merged['Longitude'], sheffield_merged['Postcode'], sheffield_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

### As the above map dosent have a key, we are going to create an output data frame which we will color with the corospoding to their cluster value. To do this we first must define the function below to return the color required for the styling of. Then we create an outputDF dataframe with the columns (Lat,long,town,region) removed to make it more readable. Then we apply the style function and output the result

```python
def highlightclusters(s):

    colorcode = rainbow[s-1]
    return 'background-color: '+colorcode
```

```python
outputDF = sheffield_merged.loc[sheffield_merged['Cluster Labels'] >= 0, sheffield_merged.columns[[0] + list(range(5, sheffield_merged.shape[1]))]].sort_values(by='Cluster Labels')
outputDF.style.applymap(highlightclusters,subset=['Cluster Labels'])
```
